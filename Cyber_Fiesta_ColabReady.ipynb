🔐 Section 3: RSA and Its Vulnerability

```python
import matplotlib.pyplot as plt

x = [64, 128, 256, 512, 1024, 2048]
classical_time = [0.01, 0.05, 0.2, 1.5, 20, 400]
quantum_time = [0.001, 0.002, 0.004, 0.008, 0.01, 0.02]

plt.plot(x, classical_time, label='Classical Factoring Time')
plt.plot(x, quantum_time, label='Quantum Factoring Time')
plt.xlabel('RSA Key Size (bits)')
plt.ylabel('Relative Time (log scale)')
plt.title('Classical vs Quantum Runtime for Factoring')
plt.legend()
plt.grid(True)
plt.show()
```

⚗️ Section 4: Qiskit Demo – Shor’s Algorithm

```python

# ✅ Using Sampler (pure Python simulator for Colab)
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

sampler = Sampler()
job = sampler.run(qc)
result = job.result()
counts = result.quasi_dists[0]

plot_histogram(counts, title="Sampler Output (No Aer Needed)")
```

🧪 Experiment: Try Different Numbers!

```python
import random
import math

def is_coprime(a, b):
    return math.gcd(a, b) == 1

def modular_check(a, N):
    r = 1
    while pow(a, r, N) != 1:
        r += 1
        if r > 1000:
            return None
    return r

a = random.randint(2, 14)
while not is_coprime(a, 15):
    a = random.randint(2, 14)

print(f"🔢 Random base a = {a}, Period r = {modular_check(a, 15)}")
```
